### Creating New Variables Conditionally (IF-THEN-ELSE)

```python
df = employees[['Country', 'Department', 'Job_Title', 'Salary']].copy()
df.head(3)
```

- Let's give a bonus to Sales people Rep. I

```python
https://datatofish.com/if-condition-in-pandas-dataframe/
    https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column
```

- If you have more than 2 conditions then use np.select

```python
conditions = [(df['Country'] == 'AU') & (df['Job_Title'] == 'Sales Rep. I'),
              (df['Country'] == 'AU') & (df['Job_Title'] == 'Sales Rep. II')]
choices = [0.1, 0.2]
df['Bonus_Rate'] = np.select(conditions, choices, default = 0)
```

- Passing Lambda Function

```python
employees.groupby(['Country', 'Gender']).agg({'Employee_ID':'count',
                                              'Dependents':['min', 'max'],
                                              'Salary': lambda x: x.max() - x.min()})
```

- You can have agg() use a custom function to be executed on specified column

```python
def percentile(n):
    def percentile_(x):
        return x.quantile(n)
    percentile_.__name__ = 'percentile_{:2.0f}'.format(n*100)
    return percentile_
```

```python
employees.groupby(['Country', 'Gender'])[['Salary']].agg([np.sum, np.mean, np.std, np.median, 
                                                          np.var, np.min, np.max, percentile(.5), percentile(.95)])
```

```python
order2 = order[['Customer_ID', 'Employee_ID', 'Order_Date', 'Delivery_Date', 'Quantity', 'Total_Retail_Price']].copy()
order3 = order[['Customer_ID', 'Employee_ID', 'Order_Date', 'Delivery_Date', 'Quantity', 'Total_Retail_Price']].copy()
```

```python
order2['Order_Date'] = order2['Order_Date'].astype('datetime64')
order2['Delivery_Date'] = order2['Delivery_Date'].astype('datetime64')
```

### Creating different Date Granularity

```python
order2['Order_Year'] = order2['Order_Date'].dt.year
order2['Order_Month'] = order2['Order_Date'].dt.month
order2['Order_Day'] = order2['Order_Date'].dt.day
order2['Order_Quarter'] = order2['Order_Date'].dt.quarter
```

```python
order2.head(3)
```

### Creating Dates (another way)

```python
order3['Order_Month'] = order3['Order_Date'].dt.to_period("M")
order3['Order_Year'] = order3['Order_Date'].dt.to_period("Y")
order3['Order_Day'] = order3['Order_Date'].dt.to_period("D")
order3['Order_Quarter'] = order3['Order_Date'].dt.to_period("Q")
```

### Get Different between 2 Dates (in days, months, years).

```python
df = times.copy()
df['Start_date'] = df['Start_date'].astype('Datetime64')
df['End_date'] = df['End_date'].astype('Datetime64')
df.head(3)
```

```python
df['diff_days'] = df['End_date'] - df['Start_date']
df['diff_in_Days']=df['diff_days']/np.timedelta64(1,'D')
df['diff_in_Weeks']=df['diff_days']/np.timedelta64(1,'W')
df['diff_in_Months']=df['diff_days']/np.timedelta64(1,'M')
df['diff_in_Years']=df['diff_days']/np.timedelta64(1,'Y')
df.head(10)
```
