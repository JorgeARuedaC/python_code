## First and Last Functions

```python
# Get last by one group, after adding salary
df = sas.sd2df(table='specialsals', libref='test')
df.sort_values(by=['Dept'], inplace=True)
df["cum_sum"] = df.groupby(['Proj', 'Dept'])['Salary'].transform(pd.Series.cumsum)
df.groupby(['Proj', 'Dept']).last()
```

```python
# Get last by two groups, after adding salary
df = sas.sd2df(table='projsals',libref='test')
df.sort_values(by=["Proj", "Dept"], inplace=True)
df["cum_sum"] = df.groupby(['Proj', 'Dept'])['Salary'].transform(pd.Series.cumsum)
df.groupby(['Proj', 'Dept']).last()
```

```python
def cel_to_farh(x):
    y = 1.8*x + 32
    return y
```

```python
# cols = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
cols = ["jan", "feb", "mar"]
new_cols = ["janf", "febf", "marf"]
df[new_cols] = df[cols].apply(cel_to_farh)
df.head(3)
```

```python
df = sas.sd2df(table='sales2',libref='sasp3')
df.head(3)
```

```python
df['Total'] = df.apply(lambda x: x['budget'] + x['actual'], axis=1)
```

```python
df['Total'] = df['Qtr1'] + df['Qtr2'] + df['Qtr3'] + df['Qtr4']
df.head(3)
```

```python
cols = ["Qtr1", "Qtr2", "Qtr3", "Qtr4"]
new_cols = ["PctQ1", "PctQ2", "PctQ3", "PctQ4"]

def get_diff(x, total):
    y = x / total
    return y

df[new_cols] = df.apply(lambda x: get_diff(x[cols], x['Total']), axis=1)
df.head(5)
```

```python
df.style.format({
    'PctQ1': '{:,.2%}'.format,
    'PctQ2': '{:,.2%}'.format,
    'PctQ3': '{:,.2%}'.format,
    'PctQ4': '{:,.2%}'.format
})
```
